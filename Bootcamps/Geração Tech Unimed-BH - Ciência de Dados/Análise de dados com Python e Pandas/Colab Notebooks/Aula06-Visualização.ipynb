# VISUALIZANDO PLANILHAS DO **EXCEL**

```python
import pandas as pd
```

```python
folder_path = "/content/drive/MyDrive/Colab Notebooks/datasets/"
```

```python
df1 = pd.read_excel(f"{folder_path}Aracaju.xlsx")
df2 = pd.read_excel(f"{folder_path}Fortaleza.xlsx")
df3 = pd.read_excel(f"{folder_path}Natal.xlsx")
df4 = pd.read_excel(f"{folder_path}Recife.xlsx")
df5 = pd.read_excel(f"{folder_path}Salvador.xlsx")
```

```python
df5.head()
```

```python
df = pd.concat([df1,df2,df3,df4,df5])
df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
df.dtypes
```

```python
df["LojaID"] = df["LojaID"].astype("object")
df.dtypes
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
df["Vendas"].mean()
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
df["Vendas"].fillna(0, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.dropna(subset=["Vendas"], inplace=True)
```

```python
df.dropna(how="all", inplace=True)
```

## **Criando colunas novas**

```python
df["Receita"] = df["Vendas"].mul(df["Qtde"])
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"] 
df.head()
```

```python
df["Receita"].max()
```

```python
df["Receita"].min()
```

```python
df.nlargest(3, "Receita")
```

```python
df.nsmallest(3, "Receita")
```

```python
df.groupby("Cidade")["Receita"].sum()
```

```python
df.sort_values("Receita", ascending=False).head(10)
```

## **Trabalhando com datas**

```python
df["Data"] = df["Data"].astype("int64")
df.dtypes
```

```python
df["Data"] = pd.to_datetime(df["Data"])
df.dtypes
```

```python
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
df["Ano_Venda"] = df["Data"].dt.year
df.sample(5)
```

```python
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
df.sample(5)
```

```python
df["Data"].min()
```

```python
df["diferenca_dias"] = df["Data"] - df["Data"].min()
df.sample(5)
```

```python
df["trimestre_venda"] = df["Data"].dt.quarter
df.sample(5)
```

```python
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
vendas_marco_19.sample(20)
```
